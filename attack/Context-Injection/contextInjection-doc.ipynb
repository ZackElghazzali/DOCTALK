# Findings and Analysis

### Summary of Findings by Defense Version
Attacks on V0: Context structure attacks had an 83.33% (12.5/15) success rate, returning information about the system's context block and user ID structure that could be used in a future context injection attack. Agent information attacks had a 100% (10/10) success rate, making it easier for an attacker to direct their injections to specific parts of the system on this version. Tool information attacks also had a 100% (5/5) success rate, giving an attacker further knowledge of the system's capabilites and potential weak points. Of the information poisoning attacks, the naive attacks proved the least effective with a 0% (0/10) success rate, displaying the defensive capabilities of the LLM's pretraining alone. Incoherence and agent denial attacks were both 100% (10/10) successful, indicating the ease of making the system useless. Lastly, harmful advice attacks had an 80% (8/10) success rate, showing that the model's pretraining can occasionally prevent the creation of harmful outputs, even with context injection.

Attacks on V1: Context structure attacks had a 0% (0/15) success rate, serving as an example of how powerful prompt hardening can be, even on its own. Agent information attacks had a 50% (5/10) success rate since prompt hardening alone can be bypassed with enough prompt engineering and iteration. Tool information, incoherence, and agent denial attacks both had 100% (5/5, 10/10) success rates, an example of the blindspots that exist when prompt hardening is used as the only defense. Once again, the naive attacks proved the least effective with a 0% (0/10) success rate. Lastly, harmful advice attacks had an 20% (2/10) success rate, an improvement over V0, but not perfect.

Attacks on V2: All attacks had a 0% success rate against V2 due to the content filtering implemented in this version. Every attack prompt developed for context injection contained some dangerous pattern or word, triggering a rejection response.

### Conclusion

Despite being one of the simplest forms of attack on an agentic system, context injections can still be incredibly dangerous and lead to an agent revealing harmful information, potentially laying the groundwork for even more devestating attacks. These attacks can also be tricky to defend since many AI systems have similar context structures, making them easy to generalize across different systems.

Prompt hardening proved to be somewhat effective against certain context injection attacks, however, it is far from a perfect defense and can usually be circumvented with enough trial and error. Given the simplicity of these attacks, it is not difficult to run many iterations of a prompt until one is successful. In order for prompt hardening alone to be effective enough to successfully defend a real system, it would need to be so restrictive that it would drastically reduce its usability.

Content filtering proved to be an incredibly effective defense against context injection since context and prompt injection attack prompts will typically need to have certain words or phrases such as, "DO NOT DELEGATE", "SYSTEM INSTRUCTION", or "IGNORE PREVIOUS INSTRUCTIONS" in order to actually be effective. This was the case in our testing as every one of our attacks was caught in the input filter without being sent to the agents.

One further defense that could be implemented to prevent context injections is context splitting, where the user's message and the context block are sent to the agents in two separate streams. This prevents the user from attempting to insert their own context into their message.
